````python
# This code was generated by ChatGPT and expanded on for more rationales.
def auto_reason(move: str) -> str:
    """
    Generate a stable, non-tactical reason for a move.
    """
    # Pawn pushing
    if move in ["e2e4", "d2d4", "e7e5", "d7d5"]:
      return "It controls the central squares of the board."

    # Castling
    if move in ["e1g1", "e8g8", "e1c1", "e8c8"]:
      return "We want to castle so that the king is not in between the rooks to improve king safety."

    # Knight development
    if move in ["b1c3", "g1f3", "b8c6", "g8f6"]:
      return "Developing one knight toward the center of the board is essential in the beginning of the game."

    # Side knight exception
    if move[2] in "ah" and abs(int(move[1]) - int(move[3]) == 2):
      return "Although we typically do not want knights and bishops on the a or h-file, we must move here for optimal piece safety or attacking."

    # Pawn promotion
    if len(move) == 5:
      return "If our pawn makes it to the other side of the board, we can promote it into another piece."

    # Default
    return "It improves the position while limiting the opponent’s counterplay."
````

````python
import re
UCI_RE = re.compile(r"([a-h][1-8][a-h][1-8][qrbn]?)")
def extract_uci(output_text: str) -> str:
    """
    Extract UCI move from XML-style or raw text.
    """
    match = UCI_RE.search(output_text)
    if not match:
        raise ValueError(f"Could not extract UCI move from: {output_text}")
    return match.group(1)
````

````python
def formatting_prompts_func(examples):
   # Construct conversations from 'instruction', 'input', and 'output'
   convos = []
   for i in range(len(examples["instruction"])): # Iterate through each example in the batch
       instruction = "Select the best move for this position given the set of legal moves"
       input_text = examples["input"][i]
       output_text = examples["output"][i] # add the output from the response function to help train

       user_content = f"{instruction} and briefly provide a reason for your choice."
       if input_text and input_text.strip(): # Only add input if it's not empty
           user_content += f"\n{input_text}"

       move = extract_uci(output_text)
       reason = auto_reason(move)
       output_text = f"{output_text}\n Reason: {reason}"

       convo = [
           {"role": "user", "content": user_content},
           {"role": "assistant", "content": output_text},
       ]
       convos.append(convo)

   texts = [tokenizer.apply_chat_template(convo, tokenize = False, add_generation_prompt = False) for convo in convos]
   return { "text" : texts, }

dataset = dataset.map(formatting_prompts_func, batched = True)
````

````python
from trl import SFTTrainer, SFTConfig

# Assuming `model` and `tokenizer` are already defined from previous steps

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    eval_dataset = None, # Can set up evaluation!
    args = SFTConfig(
        dataset_text_field = "text",
        per_device_train_batch_size = 32,
        gradient_accumulation_steps = 8, # Use GA to mimic batch size!
        warmup_steps = 10,
        max_steps = 100,
        learning_rate = 2e-5, # Reduce to 2e-5 for long training runs
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.1,
        lr_scheduler_type = "cosine",
        seed = 3407,
        report_to = "none", # Use TrackIO/WandB etc
    ),
)

# If you modify this, remember to re-run the `trainer = train_on_responses_only(...)` cell if you're using that feature.
````

### Train the model
Now let's train our model. We do 60 steps to speed things up, but you can set `num_train_epochs=1` for a full run, and turn off `max_steps=None`.

We also use Unsloth's `train_on_completions` method to only train on the assistant outputs and ignore the loss on the user's inputs. This helps increase accuracy of finetunes!

````python
from unsloth.chat_templates import train_on_responses_only
trainer = train_on_responses_only(
    trainer,
    instruction_part = "<|im_start|>user\n",
    response_part = "<|im_start|>assistant\n",
)
````

### Inference
Let's run the model via Unsloth native inference! According to the `Qwen-3` team, the recommended settings for instruct inference are `temperature = 0.7, top_p = 0.8, top_k = 20`

For reasoning chat based inference, `temperature = 0.6, top_p = 0.95, top_k = 20`

````python
messages = [
    {"role" : "user", "content" : "Select the best move for this position from the set of legal moves. Please justify your reasoning in 1 or 2 sentences. FEN: rnbqkbnr\/pppppppp\/8\/8\/4P3\/8\/PPPP1PPP\/RNBQKBNR b KQkq - 0 1\nLegal moves: g8h6, g8f6, b8c6, b8a6, h7h6, g7g6, f7f6, e7e6, d7d6, c7c6, b7b6, a7a6, h7h5, g7g5, f7f5, e7e5, d7d5, c7c5, b7b5, a7a5"}
]
text = tokenizer.apply_chat_template(
    messages,
    tokenize = False,
    add_generation_prompt = True, # Must add for generation
)

from transformers import TextStreamer
_ = model.generate(
    **tokenizer(text, return_tensors = "pt").to("cuda"),
    max_new_tokens = 30, # Increase for longer outputs!
    temperature = 0.0, top_p = 0.01, top_k = 20, do_sample=False, # For non thinking
    streamer = TextStreamer(tokenizer, skip_prompt = True),
)
````

### GGUF / llama.cpp Conversion
To save to `GGUF` / `llama.cpp`, we support it natively now! We clone `llama.cpp` and we default save it to `q8_0`. We allow all methods like `q4_k_m`. Use `save_pretrained_gguf` for local saving and `push_to_hub_gguf` for uploading to HF.

Some supported quant methods (full list on our [Wiki page](https://github.com/unslothai/unsloth/wiki#gguf-quantization-options)):
* `q8_0` - Fast conversion. High resource use, but generally acceptable.
* `q4_k_m` - Recommended. Uses Q6_K for half of the attention.wv and feed_forward.w2 tensors, else Q4_K.
* `q5_k_m` - Recommended. Uses Q6_K for half of the attention.wv and feed_forward.w2 tensors, else Q5_K.

Likewise, if you want to instead push to GGUF to your Hugging Face account, set `if False` to `if True` and add your Hugging Face token and upload location!

````python
# Save to 8bit Q8_0
if False:
    model.save_pretrained_gguf("model", tokenizer,)
# Remember to go to https://huggingface.co/settings/tokens for a token!
# And change hf to your username!
if False:
    model.push_to_hub_gguf("hf/model", tokenizer, token = "")

# Save to 16bit GGUF
if False:
    model.save_pretrained_gguf("model", tokenizer, quantization_method = "f16")
if False: # Pushing to HF Hub
    model.push_to_hub_gguf("hf/model", tokenizer, quantization_method = "f16", token = "")

# Save to q4_k_m GGUF
if False:
    model.save_pretrained_gguf("model", tokenizer, quantization_method = "q4_k_m")
if False: # Pushing to HF Hub
    model.push_to_hub_gguf("hf/model", tokenizer, quantization_method = "q4_k_m", token = "")

# Save to multiple GGUF options - much faster if you want multiple!
if False:
    model.push_to_hub_gguf(
        "hf/model", # Change hf to your username!
        tokenizer,
        quantization_method = ["q4_k_m", "q8_0", "q5_k_m",],
        token = "", # Get a token at https://huggingface.co/settings/tokens
    )
````
